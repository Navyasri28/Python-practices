For first we need to install python + pip and also MySQL and then connecting python and MySQL using MySQL connector

To install Python and Pip, begin by downloading Python from the official website, python.org. To install MySQL, start by downloading it from the official MySQL website, mysql.com. During the installation, select and install MySQL Server, MySQL Workbench, and MySQL Shell. After the installation is complete, configure MySQL by setting a root password and, if needed, create additional users to manage database access. This setup allows you to efficiently manage databases and execute SQL commands through both graphical and command-line interfaces.

To set up environment variables for MySQL, first locate the directory where MySQL is installed, typically found under "Program Files" on your C: drive. Within the installation folder, find the "bin" directory, which contains the necessary executable files. Copy the path to this "bin" directory. Next, access the system environment variables settings through the Control Panel or by searching "Edit the system environment variables" in the Start menu. In the environment variables window, find and select the "Path" variable under System variables, and click "Edit." Add the copied MySQL "bin" directory path as a new entry. This allows you to run MySQL commands from the command line without needing to specify the full path to the executable.

To verify the installations, open the command prompt and run specific commands to check the versions of MySQL, Python, and Pip. For MySQL, type `mysql --version` and press Enter; this will display the installed version of MySQL. For Python, type `python --version` to confirm the Python installation and its version. Finally, check Pip by typing `pip --version` to ensure it is installed correctly and to see its version. These commands help confirm that MySQL, Python, and Pip are properly installed and accessible from the command line.

To connect Python to MySQL, start by installing the MySQL Connector for Python using Pip. Open the command prompt and run the command `pip install mysql-connector-python`. This will download and install the necessary connector package. Next, write a simple Python script to establish the connection. Begin by importing the connector module with `import mysql.connector`. Then, create a connection object using `mysql.connector.connect()` and provide the required parameters such as host, user, and password. This script will enable Python to interact with the MySQL database.

To run the Python script for connecting to the MySQL database, follow these steps:

1)Create a new Python script file named example.py.

2)In this script, write the following code:

```python
import mysql.connector

con = mysql.connector.connect(
    host="localhost",
    user="root",
    password="your_password"
)

if con.is_connected():
    print("connection established")
```

3)Save the file.

4)Open the command prompt, navigate to the directory where example.py is saved, and run the script by typing python example.py and pressing Enter.

If the connection is successful, the script will print "connection established."
