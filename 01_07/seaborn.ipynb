# **Seaborn**

Seaborn is a popular Python library that enhances data visualization by providing a high-level interface for creating attractive and informative statistical graphics. It is built on top of Matplotlib and integrates seamlessly with Pandas dataframes, allowing users to easily visualize and analyze data. Seaborn simplifies the process of generating complex plots and offers a range of default themes and color palettes, making it a powerful tool for quickly and effectively exploring datasets.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Display the first few rows of the dataset
print(tips.head())

# Create a simple scatter plot to visualize the relationship between total bill and tip
sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.title('Scatter Plot of Total Bill vs Tip')
plt.show()

# Create a box plot to compare tips across different days of the week
sns.boxplot(x='day', y='tip', data=tips)
plt.title('Box Plot of Tips by Day')
plt.show()

# Create a histogram to visualize the distribution of total bills
sns.histplot(tips['total_bill'], bins=20, kde=True)
plt.title('Histogram of Total Bill')
plt.show()
```

**Line plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Calculate the average tip amount per day
average_tips = tips.groupby('day')['tip'].mean().reset_index()

# Create a line plot to show the average tip amount per day
sns.lineplot(x='day', y='tip', data=average_tips)
plt.title('Average Tip Amount Per Day')
plt.xlabel('Day')
plt.ylabel('Average Tip')
plt.show()
```

**Styling the lineplot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Calculate the average tip amount per day
average_tips = tips.groupby('day')['tip'].mean().reset_index()

# Set a custom style
sns.set(style="whitegrid")

# Create a line plot with custom styles
sns.lineplot(
    x='day',
    y='tip',
    data=average_tips,
    marker='o',
    linestyle='--',
    color='b',
    linewidth=2.5
)

# Customize the plot
plt.title('Average Tip Amount Per Day', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Average Tip ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

**Histogram**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Set a custom style
sns.set(style="whitegrid")

# Create a histogram to show the distribution of total bills
sns.histplot(
    tips['total_bill'],
    bins=20,
    kde=True,
    color='purple'
)

# Customize the plot
plt.title('Distribution of Total Bills', fontsize=16)
plt.xlabel('Total Bill ($)', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

**Barplot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Calculate the average total bill per day
average_bills = tips.groupby('day')['total_bill'].mean().reset_index()

# Set a custom style
sns.set(style="whitegrid")

# Create a bar plot to show the average total bill per day
sns.barplot(
    x='day',
    y='total_bill',
    data=average_bills,
    palette='viridis'
)

# Customize the plot
plt.title('Average Total Bill Per Day', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Average Total Bill ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

**Heatmap**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the flights dataset
flights = sns.load_dataset('flights')

# Pivot the dataset to get a matrix format suitable for a heatmap
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')

# Set a custom style
sns.set(style="white")

# Create a heatmap to show the number of passengers over time
plt.figure(figsize=(10, 8))
sns.heatmap(
    flights_pivot,
    annot=True,
    fmt='d',
    cmap='YlGnBu',
    linewidths=.5
)

# Customize the plot
plt.title('Number of Passengers Over Time', fontsize=16)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Month', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

**Rug plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Set a custom style
sns.set(style="whitegrid")

# Create a rug plot to show the distribution of total bills
sns.rugplot(x='total_bill', data=tips, height=0.5, linewidth=2.5)

# Customize the plot
plt.title('Rug Plot of Total Bills', fontsize=16)
plt.xlabel('Total Bill ($)', fontsize=14)
plt.ylabel('Density', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks([])  # Hide y-axis ticks

# Show the plot
plt.show()
```

**Box plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Set a custom style
sns.set(style="whitegrid")

# Create a box plot to show the distribution of tips by day
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='tip', data=tips, palette='Set3')

# Add a title and labels
plt.title('Distribution of Tips by Day', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Tip Amount ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

**Violin plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Set a custom style
sns.set(style="whitegrid")

# Create a violin plot to show the distribution of tips by day
plt.figure(figsize=(10, 6))
sns.violinplot(x='day', y='tip', data=tips, palette='muted', inner='quartile')

# Add a title and labels
plt.title('Distribution of Tips by Day', fontsize=16)
plt.xlabel('Day of the Week', fontsize=14)
plt.ylabel('Tip Amount ($)', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)

# Show the plot
plt.show()
```

**Pair plot**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the tips dataset
tips = sns.load_dataset('tips')

# Set a custom style
sns.set(style="ticks")

# Create a pair plot to visualize relationships between numerical variables
sns.pairplot(tips, hue='sex', palette='husl')

# Customize the plot
plt.suptitle('Pair Plot of Tips Dataset', fontsize=16)
plt.show()
```
